# Handling missing values

```python
dataset['age'].median()
```

```python
dataset['age'].fillna(dataset['age'].median(),inplace=True)
```

```python
print (dataset.isnull().sum())
```

```python
dataset.drop("deck",axis=1,inplace=True)
```

```python
print (dataset.isnull().sum())
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
missing_embarked = dataset['embarked'].mode()[0]
missing_embarked
```

```python
dataset['embarked'].fillna(missing_embarked,inplace = True)
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
missing_embark_town = dataset['embark_town'].mode()[0]
missing_embark_town
```

```python
dataset['embark_town'].fillna(missing_embark_town,inplace = True)
```

# Data Visualisation

```python
dataset['sex'].unique()
```

```python
freq=dataset['sex'].value_counts()
```

```python
freq
```

```python
plt.bar(freq.index,freq.values)
plt.xlabel("Sex")
plt.ylabel("Frequency")
plt.title("Total people based on gender")
plt.show()
```

```python
plt.pie(freq.values,labels=freq.index, autopct='%.2f%%')
plt.title('Percentage analysis of gender ')
plt.show()
```

# Observations
1. Majority of passenger is male
2. men ranges between 550 to 580(64.76%)
3. female ranges between 300 to 320(35.24%)


### How many survived

```python
dataset.survived.sum()
```

```python
sns.barplot(data=dataset,x='sex',y='survived')
plt.xlabel("Sex")
plt.ylabel("Survived")
plt.title("Count of survived people based on gender")
plt.show()
```

## Observation
1. More female survived than male

```python
sns.distplot(dataset.fare)
plt.xlabel("Passenger class")
plt.ylabel("Survived")
plt.title("Count of survived people based on their passenger class")
plt.show()
```

## Observation
1. 1st class passenger survived more than third class.
2. 1st class is much safer and third class are less secure

```python
sns.scatterplot(x='fare', y='survived', hue='sex',data=dataset)
```

## Observation
1. if passenger paid a higher fare, the survival rate is more.

```python
sns.violinplot(x ="sex", y ="age", hue ="survived", data = dataset, split = True)
plt.title(" Age vs survived")
plt.show()
```

## Observations
1. survival rate is good for childern.
2. high for wome in the age range of 20 to 50.
3. less for men as the age decreases.

```python
sns.barplot(data = dataset,x="embark_town", y='survived')
plt.xlabel("embark_town")
plt.ylabel("survived")
plt.title("count of survived people based on embark town")
plt.show()
```

## Observations
1. The number of passengers survived is greater in cherbourg

```python
sns.factorplot(data=dataset,x='parch',y='survived')
plt.xlabel("Parch")
plt.ylabel("Survived")
plt.title("Count of survived people based on their parch")
plt.show()
```

## Observations
1. if a passenger is alone,the survival rate is less
2. if the family size is greater than, chances of surival decreases considerebly

```python
plt.hist(dataset['parch'])
plt.title('Analysis of Parents/children aboard')
plt.xlabel('Parents/children')
plt.ylabel('count')
plt.show()
```
